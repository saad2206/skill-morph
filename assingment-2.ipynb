```python
# 1. Create variables and print them
name = "Sadman"
age = 22
favorite_color = "Blue"

print("Name:", name)
print("Age:", age)
print("Favorite Color:", favorite_color)


# 2. Create one variable of each type and verify using type()
a_int = 10
a_float = 3.14
a_str = "Hello"
a_bool = True
a_list = [1, 2, 3]
a_tuple = (4, 5, 6)
a_dict = {"key": "value"}

print(type(a_int))
print(type(a_float))
print(type(a_str))
print(type(a_bool))
print(type(a_list))
print(type(a_tuple))
print(type(a_dict))


# 3. Calculate area and perimeter of rectangle
length = 7
width = 5

area = length * width
perimeter = 2 * (length + width)

print("Area:", area)
print("Perimeter:", perimeter)


# 4. List of Bangladeshi foods
foods = ["Biryani", "Panta Bhat", "Fuchka", "Hilsha Fry", "Kacchi Biryani"]

print("First food:", foods[0])
print("Last food:", foods[-1])


# 5. List operations
numbers = [10, 5, 20, 15, 30]
numbers.append(40)  # add
numbers.remove(20)  # remove
numbers.sort()

print("Updated list:", numbers)


# 6. Dictionary about yourself
my_info = {
    "name": "Sadman",
    "age": 22,
    "country": "Bangladesh",
    "hobby": "Coding"
}

print("My Information:", my_info)


# 7. Tuple of famous place coordinates (Cox's Bazar)
location = (21.4272, 92.0058)
print("Coordinates of Cox's Bazar:", location)


# 8. Set union and intersection
colors1 = {"red", "green", "blue"}
colors2 = {"yellow", "blue", "black"}

print("Union:", colors1.union(colors2))
print("Intersection:", colors1.intersection(colors2))


# 9. Function to print first and last character of a string
def first_last_char(text):
    print("First character:", text[0])
    print("Last character:", text[-1])

first_last_char("Bangladesh")


# 10. Function to convert string to uppercase
def make_upper(text):
    return text.upper()

print(make_upper("hello world"))
```
